## Sucesión de Fibonacci

Cálcula el n-ésimo término de la [Sucesión de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci) mediante un algoritmo recursivo y otro iterativo. Representa gráficamente cómo crece el tiempo de cómputo en función del número de términos para ambos algoritmos.

```python
#Sucesión_de_Fibonacci


import matplotlib.pyplot as plt # lib para gráficas
import time # lib para medit tiempos de ejecución

#fibonacci recursivo
def f_r(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return f_r(n - 1) + f_r(n - 2)



#fibonacci dinámico
def f_d(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1

    ultimo = 1
    actual = 1
    for i in range(n-2):
        aux = actual
        actual += ultimo
        ultimo = aux
    return actual



#acá Funciones para gráficas
x=[]
y=[]
n= 30 # Calcular hasta n-ésimo término
for i in range (1,n):
    t1 = time.time_ns()
    f_d (i)
    t2 = time.time_ns()
    x.append(i)
    y.append((t2-t1))
plt.plot(x,y, c="r")

x=[]#borrar valores para otra gráfica
y=[]#borrar valores para otra gráfica
for i in range (1,n):
    t1 = time.time_ns()
    f_r (i)
    t2 = time.time_ns()
    x.append(i)
    y.append((t2-t1)) # dividir por 1000000000 para segundos... no afecta la gráfica
plt.plot(x,y, c="b")




plt.legend(['Dinámico', 'Recursivo'])
plt.show()


```

## Devolución de cambio por técnica voraz

Resuelve el [Problema del Cambio de Moneda](https://es.wikipedia.org/wiki/Problema_de_cambio_de_monedas) utilizando una técnica voraz.

```python
# Problema del cambio de moneda
import matplotlib.pyplot as plt

def imprimir_denominaciones(resultado):
    denominaciones = list(resultado.keys())
    cantidad_monedas = list(resultado.values())


    print("Número total de monedas necesarias:",sum(cantidad_monedas), "Distribuidas de esta forma: ", resultado)
    # Extraer las claves y los valores del diccionario


    # Crear la gráfica de barras
    plt.bar(denominaciones, cantidad_monedas, color='NAVY')
    for i, valor in enumerate(cantidad_monedas):
        plt.text(denominaciones[i], valor + 0.1, str(valor), ha='center')
    # Establecer solo las denominaciones presentes en el eje x
    plt.xticks(denominaciones)

    # Añadir etiquetas y título
    plt.xlabel('Denominación')
    plt.ylabel('Cantidad de Monedas')
    plt.title('Cantidad de Monedas por Denominación')
    # Ocultar el eje y
    plt.gca().axes.get_yaxis().set_visible(False)
    # Mostrar la gráfica
    plt.show()



def cambio_de_moneda(monto, denominaciones):
    # Ordenar las denominaciones en orden descendente
    #requerido para buen funcionamiento
    denominaciones.sort(reverse=True)
    # Inicializar el contador de monedas
    total_monedas = 0
    # Inicializar diccionario, para saber cuantas monedas de cada denominación
    dicc_denominaciones={}

    # Iterar sobre cada denominación
    for d in denominaciones:
        # Calcular la cantidad de monedas de esa denominación que se pueden usar
        cantidad_monedas = monto // d
        # Actualizar el monto restante
        monto -= cantidad_monedas * d
        # Incrementar el contador de monedas utilizadas
        total_monedas += cantidad_monedas
        dicc_denominaciones[d] = cantidad_monedas


    return dicc_denominaciones


monto = 62
denominaciones = [1, 5, 10, 20,50]  # Denominaciones de moneda
resultado = cambio_de_moneda(monto, denominaciones)
imprimir_denominaciones(resultado)





```

## N-Reinas por técnica de vuelta atrás
Resuelve el [Problema de las N-Reinas](https://es.wikipedia.org/wiki/Problema_de_las_ocho_reinas) en un tablero de dimensión N mediante la técnica de la vuelta atrás (backtraking).

```python
def imprimir_tablero(TuplaQ,n):
    print("")
    print("Solución encontrada:")

    tablero = [[' '] * n for _ in range(n)]
    r=0
    for T in TuplaQ:
        #print(int(T),r)
        tablero[int(T)][r]="Q"
        r=r+1
    for fila in tablero:
        print(fila)
        print("")

import matplotlib.pyplot as plt


def imprimir_tablero_Jupyter(TuplaQ,n):
    # Crear una figura y un eje
    fig, ax = plt.subplots()

    # Dibujar los cuadros del tablero
    for i in range(n):
        for j in range(n):
            # Si la suma de las coordenadas es par, el cuadro será blanco; de lo contrario, será gris
            color = 'white' if (i + j) % 2 == 0 else '#f0f0f0'  # Color gris claro en formato hexadecimal
            # Dibujar un cuadrado en la posición (i, j) con el color correspondiente
            ax.add_patch(plt.Rectangle((i, j), 1, 1, color=color))

    # Agregar círculos con la letra "Q" para representar las reinas en los cuadros seleccionados
    #posiciones_reinas= [(0, 2), (1, 0),(2, 3),(3, 1)]
    # debo cambiar la Tupla [1, 3, 0, 2] por posiciones_reinas con ese formato: [(0,2 ), (2, 1)]
    m=n-1
    posiciones_reinas = [(i, m-j) for i, j in enumerate(TuplaQ)]
    #print(posiciones_reinas)
    for pos in posiciones_reinas:
        x, y = pos
        # Dibujar un círculo negro
        circle = plt.Circle((x + 0.5, y + 0.5), 0.4, color='black', fill=True)
        ax.add_artist(circle)
        # Agregar la letra "Q"
        ax.text(x + 0.5, y + 0.5, 'Q', color='white', ha='center', va='center', fontsize=20)

    # Establecer límites del eje y aspecto
    ax.set_xlim(0, n)
    ax.set_ylim(0, n)
    ax.set_aspect('equal')

    # Ocultar los ejes
    ax.axis('off')

    # Mostrar el tablero
    plt.show()



def es_seguro(TuplaQ,columna,i):
    # Verifica si una reina puede ser colocada en posición
    # Verifica la fila
    #DOS REINAS ESTARAN EN LA MISMA FILA SI HAY DOS VALORES IGUALES  EJEMPLO [1,1,*,*]
    if i in TuplaQ:
        return False

    # Verifica la diagonal
    #DOS REINAS ESTARAN EN LA MISMA DIAGONAL SI |Xi - Xj| = |i-j|
    # EJEMPLO: se intenta acomodar una reina en la segunda columna fila 2; entoces
    # [1,¿2?,' ',' '] -> |2-1| = |2-1|
    # EJEMPLO [1,*,*,4] -> |4-1| = |4-1|

    #NO estan en diagonal:
    # EJEMPLO [1,3,*,*] -> |2-1| = |3-1|
    for j in range(0,n):

        if columna!=j: #si son diferentes, ya que no voy a preguntar por la misma columna
            #print( i,j)
            #print(TuplaQ[i],TuplaQ[j])
            T1 =  i   # es el nuevo valor que se intenta
            T2 = int(TuplaQ[j]) if TuplaQ[j] != " " else n*2 # Tocó poner este valor lo suficientemente grande para que nolo tome como seguro
            #print( columna,j, T1,T2)
            if abs(columna-j) == abs(T1-T2) :
                #Si entra por acá es por que encontró una reina existente en diagonal
                return False


    return True

def resolver_reinas_aux(TuplaQ,columna):
    ######print(TuplaQ)  # si se quiere ver los caminos parciales antes de la solución
    # Caso base: si todas las reinas han sido colocadas, devuelve True
    if columna >= n:
        return True

    # Recorre cada fila en la columna actual
    for i in range(0,n):
        # Verifica si es seguro colocar una reina en esta posición
        if es_seguro(TuplaQ, columna,i):
            TuplaQ[columna]=i
            # Llamada recursiva para colocar las reinas restantes
            if resolver_reinas_aux(TuplaQ, columna + 1):
                return True
            # Si colocar la reina en esta posición no lleva a una solución, retrocede (backtrack)
            TuplaQ[columna]=" "

    # Si ninguna posición es segura en esta columna, devuelve False
    return False


def resolver_reinas(n):

    TuplaQ=[] #para guardar la posición de las reinas
    for i in range(0, n):
        TuplaQ.append(' ')
    #print(TuplaQ)


    # Llama a la función auxiliar para resolver el problema
    if not resolver_reinas_aux(TuplaQ,0):
        print("No encontró solución...")
        return False

    print(TuplaQ)

    # Imprime el tablero con las reinas colocadas
    #imprimir_tablero(TuplaQ,n) #  imprimir en consola
    imprimir_tablero_Jupyter(TuplaQ,n) #  imprimir en imagen (para jupyter)
    return True




# Llama a la función principal para resolver el problema de las n reinas
n=10
resolver_reinas(n)

```
